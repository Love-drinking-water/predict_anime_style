```python
# 导入所需的模块
import numpy as np # numpy数组
import pandas as pd # pandas表格
import matplotlib.pyplot as plt # 数据可视化
import matplotlib.font_manager as mfm # 设置字体
import openpyxl # 用于读取表格的包
plt.style.use('ggplot') # 使用ggplot风格
plt.rcParams['axes.unicode_minus'] = False# 用来正常显示负号
```

```python
data = pd.read_excel(
    '../data/动画信息.xlsx',
     engine='openpyxl') # 读取数据
np.random.seed(233) # 设置随机种子
```

```python
data.columns
```

```python
dict1 = {
    '风格':['搞笑', '百合',
       '治愈', '后宫', '校园', '恋爱', '科幻', '日常', '战斗', '热血'],
    'counts':[1941,793,1037,683,1037,994,1114,868,1493,925]
}
counts = pd.DataFrame(dict1) # 建立风格-出现次数的关系
```

```python
font_path = "../data/simhei.ttf" # 字体路径
prop = mfm.FontProperties(fname=font_path,size=25) # 自定义字体（大字号）
prop_smallsize = mfm.FontProperties(fname=font_path,size=20) # 自定义字体（小字号）
x = ['搞笑','百合','治愈','后宫','校园','恋爱','科幻','日常','战斗','热血']
plt.figure(figsize=(10, 7)) # 设置画布大小
plt.bar(x=range(len(x)), height=counts['counts'],color="pink") # 绘画条形图
plt.title("对应风格的动画数量",font_properties=prop) # 绘画标题
plt.xticks(range(len(x)),x,fontproperties=prop_smallsize) # 绘画横轴坐标
plt.yticks(range(0,2001,250),fontsize=20) # 纵轴坐标
plt.xlabel('风格',font_properties=prop) # 横轴标签
plt.ylabel('数量',font_properties=prop) # 纵轴标签
plt.show() # 展示结果
```
